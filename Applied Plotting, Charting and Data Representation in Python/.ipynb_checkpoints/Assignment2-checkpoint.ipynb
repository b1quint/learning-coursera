```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

filename = ("./data/temps.csvs.csvps.csv")

# 'ID', 'Date', 'Element', 'Data_Value'
df = pd.read_csv(filename)
df.Date = pd.to_datetime(df.Date, format='%Y-%m-%d')

df['year'] = df.Date.dt.year
df['month'] = df.Date.dt.month
df['day'] = df.Date.dt.day

df_tmax = df.query('Element == "TMAX" and Date < "2015-01-01"')
group_tmax = df_tmax.groupby(by=['month', 'day'])
max_tmax = group_tmax.Data_Value.max().to_frame().reset_index()
max_tmax['year'] = 2000
max_tmax['Date'] = pd.to_datetime(max_tmax[['year', 'month', 'day']], format="%Y-%m-%d")
max_tmax = max_tmax.rename(columns={'Data_Value': 'record'})

df_tmin = df.query('Element == "TMIN" and Date < "2015-01-01"')
group_tmin = df_tmin.groupby(by=['month', 'day'])
min_tmin = group_tmin.Data_Value.min().to_frame().reset_index()
min_tmin['year'] = 2000
min_tmin['Date'] = pd.to_datetime(min_tmin[['year', 'month', 'day']], format="%Y-%m-%d")
min_tmin = min_tmin.rename(columns={'Data_Value': 'record'})

tmax2015 = df.query('Element == "TMAX" and Date >= "2015-01-01"')
gtmax2015 = tmax2015.groupby(by=['month', 'day'])
rtmax2015 = gtmax2015.Data_Value.max().to_frame().reset_index()
rtmax2015 = rtmax2015.rename(columns={'Data_Value': 'record_2015'})

cmax = lambda s: s.record_2015 if s.record_2015 > s.record else None

rtmax = pd.merge(max_tmax, rtmax2015, on=['month', 'day'])
rtmax = rtmax[['Date', 'year', 'month', 'day', 'record', 'record_2015']]
rtmax['broken_record'] = rtmax.apply(cmax, axis=1)

tmin2015 = df.query('Element == "TMIN" and Date >= "2015-01-01"')
gtmin2015 = tmin2015.groupby(by=['month', 'day'])
rtmin2015 = gtmin2015.Data_Value.min().to_frame().reset_index()
rtmin2015 = rtmin2015.rename(columns={'Data_Value': 'record_2015'})

cmin = lambda s: s.record_2015 if s.record_2015 < s.record else None

rtmin = pd.merge(min_tmin, rtmin2015, on=['month', 'day'])
rtmin = rtmin[['Date', 'year', 'month', 'day', 'record', 'record_2015']]
rtmin['broken_record'] = rtmin.apply(cmin, axis=1)

fig, ax = plt.subplots(num="Temperature Records", figsize=(10, 6))

ax.plot(rtmax.record, 'C3-', label='High record temperature', alpha=0.3)
ax.plot(rtmin.record, 'C0-', label='Low record temperature', alpha=0.3)

ax.scatter(rtmax.index, rtmax.broken_record, marker='v', c='C3', 
           label='High record temperature broken in 2015')

ax.scatter(rtmax.index, rtmin.broken_record, marker='^', c='C0', 
           label='Low record temperature broken in 2015')

ax.fill_between(
    rtmax.index, rtmax.record, rtmin.record,
    facecolor='black', alpha=0.1)

ax.grid(':', alpha=0.2)
ax.set_xlabel('Days over the year', fontsize='large')
ax.set_ylabel('Tenths of $^{\circ}$C', fontsize='large')
ax.tick_params(axis=u'both', which=u'both',length=0)

ax.set_xticks(major_ticks.index)
ax.set_xticklabels(major_ticks.dt.strftime('%b'))
ax.legend(loc='lower center', fancybox=None, framealpha=0, ncol=1)

ax.set_title('High and Low Temperature Records per day\n between 2005 and 2014 near Ann Arbor, Michigan State')

for loc in ['bottom', 'top', 'right', 'left']:
    ax.spines[loc].set_visible(False)

plt.savefig('courser_assignment2_bq.png')
plt.show()
```
