In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])

df    
```

```python
%matplotlib notebook
import matplotlib as mpl
import numpy as np
import scipy.stats
import time

from matplotlib import patches
from matplotlib import pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable


class InteractiveDataVisualize:
    
    def __init__(self, _df):
        self.df = _df
        
        idx = self.df.index
        avg = self.df.apply(np.mean, axis=1)
        conf = self.df.apply(self.mean_confidence_interval, axis=1)
        
        self.idx, self.avg, self.conf = idx, avg, conf
        self.fig, self.ax, self.bars, self.cmap = self.plot_bars(idx, avg, conf)
        self.y = None
        
        def _onclick(event):
            ix, iy = event.xdata, event.ydata
            self.update_colors(iy)
        
        cid = self.fig.canvas.mpl_connect('button_press_event', _onclick)

    @staticmethod
    def mean_confidence_interval(data, confidence=0.95):
        """
        Calculates the confidence interval.

        Args:
            data (array): 
                inputs data (each row of our dataset).
            confidence (float): 
                confidence interval [0, 1].

        Returns:
            float: the value representing the confidence interval.
        """
        a = 1.0 * np.array(data)
        n = len(a)
        m, se = np.mean(a), scipy.stats.sem(a)
        h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
        return h

    def plot_bars(self, index, average, conf_interval):
        """
        Creates a bar plot that represents the average of the values for 
        each year with 95% confidence interval.
        
        Args:
            index (array): 
                index of where the bars will be plot.
            average (array): 
                average value of each row in the dataset.
            conf_interval (array): 
                95% confidence interval around the mean for each row in the dataset.
        """    
        fig, ax = plt.subplots(num='My Bar Plot')
        fig.suptitle("Probability of my data be under the dashed line", fontsize=16)

        bars = ax.bar(index, average, yerr=conf_interval, capsize=10, color="lightgray")

        _ = [ax.spines[s].set_visible(False) for s in ['top', 'right']]
        ax.set_ylabel("Value [??]")
        ax.set_xlabel("Row [year]")
        ax.set_xticks(index)
        ax.set_xticklabels(index)
        ax.grid(alpha=0.5, axis='y', color='k', lw=0.1)
        ax.set_title('Click on plot to set dashed line position.')
        
        cmap = plt.cm.coolwarm_r
        cmaplist = [cmap(i) for i in range(cmap.N)]
        cmap = mpl.colors.LinearSegmentedColormap.from_list('Custom Map', cmaplist, cmap.N)
        
        bounds = np.linspace(0, 1, 11)
        norm = mpl.colors.BoundaryNorm(bounds, cmap.N)
        
        divider = make_axes_locatable(ax)
        cax = divider.append_axes("right", size="5%", pad=0.05)

        sm = mpl.cm.ScalarMappable(cmap=cmap, norm=norm)
        sm.set_array([])
        
        cbar = plt.colorbar(sm, cax=cax)
        cbar.set_label('Probability of data be under y-line', rotation=270, labelpad=15)
        
        return fig, ax, bars, cmap
    
    def get_probability(self, ylim):
        """
        Get the discretized probability of the data be smaller than ylim.
        
        Args:
            ylim (float)
            
        Returs:
            array
        """
        top = self.avg + self.conf
        bot = self.avg - self.conf        
        
        prob = np.array((ylim - bot) / (top - bot))
        prob[prob > 1] = 1
        prob[prob < 0] = 0

        bins = np.linspace(0, 1, 11)
        binned_prob = np.digitize(prob, bins, right=True)
        return bins[binned_prob]

    
    def update_colors(self, ylim):
        """
        Update bar colors based on how much data is covered by the horizontal line at ylim.
        
        Args:
            ylim (float) 
        """        
        prob = self.get_probability(ylim)
        _ = [b.set_color(self.cmap(p)) for b, p in zip(self.bars, prob)]
            
        if self.y is None:
            self.y = self.ax.axhline(ylim, color="black", alpha=0.5, ls='--')
        else:
            self.y.set_data(([0, 1], [ylim, ylim]))        
        
        self.fig.canvas.draw()        
        
        
dataviz = InteractiveDataVisualize(df)
```
